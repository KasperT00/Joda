Koodin toteuttaminen olisi varmasti ollut helpompaa vasta demo-session jälkeen, mutta koen kuitenkin oppineen huomattavasti enemmän tehtyäni sen itse. 

```python
import scrapy
class Pasta(scrapy.Spider):    
    name = "Pasta"     
    start_urls = [
        'https://www.k-ruoka.fi/kauppa/tuotehaku/hedelmat-ja-vihannekset', 
    ]
    def parse(self, response):     #Prosessoi ja palauttaa raavitun datan
        for tuote in response.css('li.bundle-list-item'):  # Sivun tuotteiden osoite
            texti = tuote.css('span::text').get()# Tämä on siellä tuotetiedon sisällä ja se on se teksti osa
            yield {'tuote':texti} 
```

Yllä esitetty "Pasta"-ryömiijä voi kerätä tietoa sille annetuista osoitteista, mutta eri sivujen elementtien tiedot on sille kerrottava. Luomani ryömijä ei kuitenkaan pysty lukemaan esimerkiksi esimerkissä olevaa sivua kokonaan, sillä se on dynaaminen ja latautuu samalla sivua selatessa. Aion kuitenkin löytää tähän ratkaisun ja saada ryömijä toimimaan myös dynaamisilla sivuilla, sekä menemään seuraaville sivuille.

```python
import nltk
import csv
import pandas as pd
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
```

```python
arviot = ""
with open("arvostelutA10.csv", 'r',encoding='utf-8') as tiedosto:
    lukija = csv.reader(tiedosto, delimiter=',')
    next(lukija)
    for rivit in lukija:
        if(rivit[2] != '-'):
            arviot = arviot + (rivit[2])
```

```python
arvio=word_tokenize(arviot)
sallitut=set(stopwords.words("finnish"))
putsatut =[]
for sana in arvio:
    if sana not in sallitut:
        putsatut.append(sana)
putsatut =[sana.lower() for sana in putsatut if sana.isalpha()]
fd = FreqDist(putsatut)
fd.plot(30,cumulative=False)
plt.show()
```

Ohjaamaton oppiminen voi olla esimerkiksi ostoskorianalyysiä, verkostoanalyysiä, aihemallinnusta tai ryvästämistä. Kiinnostava harjoitustyön aihe olisi ollut ostoskorianalyysi. Ohjaamattoman oppimisen työkalu KMeans laskee klustereiden määrän ja jokaiselle klusterille eli osa alueelle lasketaan keskipisteet, joiden etäisyys kaikkien havaintojen keskipisteistä lasketaan. Havainnot sijoitetaan klusteriin sen perusteella, minkä alueen keskipiste on lähimpänä ja prosessia toistetaan kunnes keskipisteet eivät muutu. Tässä estimooidaan kuitenkin itse klustereiden määrä. Aiemmin käsitellyt aihemallinnus ja luonnollisen kielen analyysi ovat myös ohjaamatonta oppimista. Aihemallinnuksessa ennustetaan ja peilataan tekstin luonnetta, jossa inhmillinen tekijä on toisaalta ongelmana. Tarkoitus kuitenkin tutkia itse ihmillisyyttä ja mallinnus on täten subjektiivista. KMeans oli itselle hieman vaikea ainakin koodauksen puolesta. Tutkin myös KMeans menetelmää matemaattisesta näkökulmasta ja pienen vilkaisun jälkeen huomasin sen haastavuuden matemaattisesti. Revoledu kuitenkin selittää hyvin matemaattisen tavan laskemiselle (Revoledu 2017). 

Alle olen luonut koodin, jossa käytetään KMeans klusterointia ja laskemista. Käytin hyväksi internetin antoa KMeans klusteroinnista, sillä osaaminen ei riittänyt tarpeeksi itse dokumentoinnin lukemiseen ja tekemiseen (DatatoFish 2020). Olen käyttänyt datana Kagglen antia ja ottanut datasetin NBA pelaajista (Kaggle, linkki lähteissä). Käytetään klusteroinnissa pelaajien pituuksia ja syntymävuosia.

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

daata = pd.read_csv("C:/Users/kaspe/OneDrive - TUNI.fi/Jupyter kansio/pelaajat2.csv", na_values='-')
df = DataFrame(daata,columns=['born','height'])
df.fillna(0,inplace=True)
df = df[(df.T != 0).any()]
df.dropna()

kmeans = KMeans(n_clusters=3).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)
plt.scatter(df['born'], df['height'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

Koodin piirroksesta nähdään, kuinka KMeans on luonut kolme klusteria, kun koodille annettiin klustereiden määräksi kolme. Nämä rajautuvat hyvin eri vuosiin. Itse visuaalisesta näytöstä en itse ainakaan saa paljoa relevanttia informaatiota, mutta itse klusterointi on hienosti totuetunut.
